# 2-Sample T-Test for Working Day Effect:





```python
from scipy.stats import ttest_ind, levene

# Based on the 2-Sample T-Test:

# Null Hypothesis (H0): The mean number of electric cycles rented on working days is equal to the mean number on non-working days.
# Alternate Hypothesis (H1): The mean number of electric cycles rented on working days is different from the mean number on non-working days.

# Extracting data for working and non-working days
working_day_data = df[df['workingday'] == 1]['count']
non_working_day_data = df[df['workingday'] == 0]['count']

# Assumption Check: Normality
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.histplot(working_day_data, kde=True, color='skyblue', label='Working Day')
sns.histplot(non_working_day_data, kde=True, color='salmon', label='Non-Working Day')
plt.title('Distribution of Electric Cycles Rented')
plt.legend()


# Assumption Check: Equal Variance
statistic, p_value = levene(working_day_data, non_working_day_data)
print(f'Levene\'s Test Statistic: {statistic}, p-value: {p_value}')

# Perform 2-Sample T-Test
t_statistic, p_value_ttest = ttest_ind(working_day_data, non_working_day_data, equal_var=True)

# Set a significance level (alpha) typically at 0.05.

alpha = 0.05
if p_value_ttest < alpha:
    decision = "Reject H0"
else:
    decision = "Fail to Reject H0"

# Display Results
print(f'T-Test Statistic: {t_statistic}, p-value: {p_value_ttest}')
print(f'Decision: {decision}')

# Inference
if decision == "Reject H0":
    print("There is a significant difference in the mean number of electric cycles rented between working and non-working days.")
else:
    print("There is no significant difference in the mean number of electric cycles rented between working and non-working days.")
```

### Hypothesis Testing Results:

1. **Levene's Test for Equal Variances:**
   - Levene's Test Statistic: 0.00497
   - p-value: 0.94378

   - **Conclusion:** Fail to reject the null hypothesis of equal variances.

2. **2-Sample T-Test for Mean Difference:**
   - T-Test Statistic: 1.2096
   - p-value: 0.2264

   - **Decision:** Fail to reject the null hypothesis.

### Inference:

Based on the results of Levene's test and the 2-sample t-test:

- The Levene's test indicates that there is no significant evidence to suggest unequal variances between the number of electric cycles rented on working and non-working days.

- The 2-sample t-test results in a p-value of 0.2264, which is greater than the significance level of 0.05. Therefore, we fail to reject the null hypothesis.

### Conclusion:

There is **no significant difference** in the mean number of electric cycles rented between working and non-working days. The data does not provide enough evidence to support the claim that the mean counts are different on these two types of days.


# ANNOVA to check if No. of cycles rented is similar or different in different weather and season:


```python
from scipy.stats import f_oneway
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
# One-Way ANOVA (Weather):
# H0: The mean number of electric cycles rented is equal across different weather conditions.
# H1: The mean number of electric cycles rented is different across at least one pair of weather conditions.

# One-Way ANOVA (Season):
# H0: The mean number of electric cycles rented is equal across different seasons.
# H1: The mean number of electric cycles rented is different across at least one pair of seasons.

# Two-Way ANOVA (Interaction between Weather and Season):
# H0: The interaction between weather conditions and seasons has no significant effect on the mean number of electric cycles rented.
# H1: The interaction between weather conditions and seasons has a significant effect on the mean number of electric cycles rented.

# One-Way ANOVA
# Extracting data for different weather conditions and seasons
weather_groups = [df[df['weather'] == i]['count'] for i in range(1, 5)]
season_groups = [df[df['season'] == i]['count'] for i in range(1, 5)]

# Perform One-Way ANOVA
f_statistic_weather, p_value_anova_weather = f_oneway(*weather_groups)
f_statistic_season, p_value_anova_season = f_oneway(*season_groups)

# Display Results
print(f'One-Way ANOVA (Weather) - F-Statistic: {f_statistic_weather}, p-value: {p_value_anova_weather}')
print(f'One-Way ANOVA (Season) - F-Statistic: {f_statistic_season}, p-value: {p_value_anova_season}')

# Decision Rule
alpha = 0.05
decision_weather = "Reject H0" if p_value_anova_weather < alpha else "Fail to Reject H0"
decision_season = "Reject H0" if p_value_anova_season < alpha else "Fail to Reject H0"

print(f'Decision (Weather): {decision_weather}')
print(f'Decision (Season): {decision_season}')

# Two-Way ANOVA

# Creating a formula for Two-Way ANOVA
formula = 'count ~ C(weather) * C(season)'
model = ols(formula, df).fit()

# Performing Two-Way ANOVA
anova_table = anova_lm(model, typ=2)

# Display Results
print(anova_table)

# Decision Rule
interaction_p_value = anova_table.loc['C(weather):C(season)', 'PR(>F)']
decision_interaction = "Reject H0" if interaction_p_value < alpha else "Fail to Reject H0"

print(f'Decision (Interaction): {decision_interaction}')
```

# One-Way ANOVA Results:

## Weather:
- F-Statistic: 65.53
- p-value: 5.48e-42
- Decision: Reject H0
  - There is a significant difference in the mean number of electric cycles rented across different weather conditions.

## Season:
- F-Statistic: 236.95
- p-value: 6.16e-149
- Decision: Reject H0
  - There is a significant difference in the mean number of electric cycles rented across different seasons.

# Two-Way ANOVA Results:

## Interaction (Weather and Season):
- F-Statistic: 3.08
- p-value: 5.15e-03
- Decision: Reject H0
  - There is a significant interaction effect between weather conditions and seasons on the mean number of electric cycles rented.



#Chi-square test to check if Weather is dependent on the season

```python
from scipy.stats import chi2_contingency

## Null Hypothesis (H0):
  #The distribution of weather conditions is independent of the distribution of seasons.

## Alternative Hypothesis (H1):
  #The distribution of weather conditions is dependent on the distribution of seasons.

# Create a contingency table
contingency_table = pd.crosstab(df['weather'], df['season'])

# Perform Chi-square test
chi2_stat, p_value_chi2, dof, expected = chi2_contingency(contingency_table)

# Display Results
print(f'Chi-square Statistic: {chi2_stat}')
print(f'p-value: {p_value_chi2}')
print(f'Degrees of Freedom: {dof}')

# Set a significance level (alpha)
alpha = 0.05

# Decision Rule
if p_value_chi2 < alpha:
    print("Reject H0--There is a significant association between Weather and Season.")
else:
    print("Fail to Reject H0--There is no significant association between Weather and Season.")
```

## Test Results:
- Chi-square Statistic: 49.16
- p-value: 1.55e-07
- Degrees of Freedom: 9

## Decision:
- Reject H0

## Inference:
- There is a significant association between weather conditions and seasons.
